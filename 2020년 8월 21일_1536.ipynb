### 1,185.60 환율로 원화 금액으로 연봉 계산하기 (연봉_원화)

```python
kbo_salary = kbo[~kbo['연봉'].isnull()].copy()
```

```python
def func(x):
    if x.find('달러') >= 0:
        return int(x.replace("달러", "")) * 1185.6 / 10000
    else:
        return int(x.replace("만원", ""))
```

```python
kbo_salary['연봉_원화'] = kbo_salary['연봉'].apply(func)
```

```python
kbo_salary.sort_values(['구단', '연봉_원화'], ascending=[True, False]).groupby('구단').last()
```

### 각 구단 평균 연봉

```python
kbo_salary.groupby(['구단'], as_index=False)[['연봉_원화']].agg(['mean', 'min', 'max', 'median'])
```
