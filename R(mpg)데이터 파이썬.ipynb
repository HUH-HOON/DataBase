```python
welfare.rename(columns={'h10_g3' : 'gender',
                    'h10_g4' : 'birth',
                   'h10_g10' : 'marriage',
                  'h10_g11' : 'religion',
                  'p1002_8aq1' : 'income',
                   'h10_eco9' : 'code_job',
                  'h10_reg7' : 'code_region'}, inplace=True)
```

```python
welfare['gender'] = welfare['gender'].apply(lambda x : 'male' if x == 1 else 'female'  )
```

```python
welfare['gender'].value_counts()
```

```python
welfare['gender']
```

```python
welfare.groupby('gender')[['income']].mean()
```

```python
welfare['birth']
```

```python
welfare['age']=welfare['birth'].apply(lambda x: 2015 - x +1)
```

```python
welfare['age']
```

```python
welfare['aged'] = welfare['age'].apply(lambda x : 'young' if x <30 else 'middle' if x<=59 else 'old')
```

```python
welfare[['age','aged']]
```

```python
welfare.groupby('age', as_index = False)[['income']].mean().dropna()
```

```python
welfare.groupby('aged', as_index = False)[['income']].mean().dropna()
```

```python
welfare['group_marriage'] = \
            welfare['marriage'].apply(lambda x :\
                          'marriage' if x == 1 else 'divorce' if x==3 else None)
```

```python
welfare['group_marriage'].value_counts()
```
